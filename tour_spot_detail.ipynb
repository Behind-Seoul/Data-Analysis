# #3. 문화/전시 관광 경로

```python
# 3.1. 종로구 인기관광지
tour_spot_popular = pd.read_csv('data/종로구_인기관광지.csv', encoding='cp949')
tour_spot_popular = tour_spot_popular[tour_spot_popular['분류'].str.contains('전시')]
del tour_spot_popular['합산 검색 수']
tour_spot_popular.head(15)
```

```python
# 3.1. 종로구 중심관광지
tour_spot_main = pd.read_csv('data/종로구_중심관광지.csv', encoding='cp949')
tour_spot_main = tour_spot_main[tour_spot_main['분류'].str.contains('문화')]
tour_spot_main.head(15)
```

```python
# 3.2. 전처리 : 주소 -> 경위도 좌표(Kakao API 이용)
import requests

# 3.2.1 인기 관광지 변환
address = tour_spot_popular['주소'].tolist()
mapX = []
mapY = []

for i in range(len(address)):
    REST_API_KEY = '80b269050cd58c9743d68720ddc84692'

    headers = {
        'Authorization': f"KakaoAK {REST_API_KEY}",
        'Content-Type': 'application/x-www-form-urlencoded',
    }

    data = f'query={address[i]}'

    response = requests.get('https://dapi.kakao.com/v2/local/search/address.json', headers=headers, data=data.encode('utf-8'))
    text = response.json()

    mapX.append(text['documents'][0]['x'])
    mapY.append(text['documents'][0]['y'])

tour_spot_popular['lat'] = mapY
tour_spot_popular['lon'] = mapX

tour_spot_popular.to_csv('data/tour_spot_popular.csv', encoding='utf-8', index=False)
tour_spot_popular.head(15)
```

```python
# 3.2.2 중심 관광지 변환
from cmath import nan


address = tour_spot_main['주소'].tolist()
mapX = []
mapY = []

for i in range(len(address)):
    REST_API_KEY = '80b269050cd58c9743d68720ddc84692'

    headers = {
        'Authorization': f"KakaoAK {REST_API_KEY}",
        'Content-Type': 'application/x-www-form-urlencoded',
    }

    data = f'query={address[i]}'

    response = requests.get('https://dapi.kakao.com/v2/local/search/address.json', headers=headers, data=data.encode('utf-8'))
    text = response.json()

    try:
        mapX.append(text['documents'][0]['x']) 
    except Exception:
        mapX.append(nan)

    try:
        mapY.append(text['documents'][0]['y'])
    except Exception:
        mapY.append(nan)

tour_spot_main['lat'] = mapY
tour_spot_main['lon'] = mapX

tour_spot_main.dropna()
tour_spot_main.to_csv('data/tour_spot_main.csv', encoding='utf-8', index=False)
tour_spot_main.head(15)
```

```python
# 3.3 인기, 중심 관광지 병합
tour_spot_merge = pd.concat([tour_spot_main, tour_spot_popular])
tour_spot_merge.to_csv('data/tour_spot_merge.csv', encoding='utf-8', index=False)
tour_spot_merge.head()
```
